## Focus on differences

There are essentially three numeric data types in python: integer, floating point and complex numbers. There is no distinction between int and long, float and double

```python
# Most interesting: No INT_MAX
a = 1234567890
b = a * a * a
c = b * b * b
d = c * c * c 
e = d * d * d
f = e * e * e
print(f)
```

```python
# There are two divisions
a = 19
b = 3
print(a/b)
print(a//b)
```
